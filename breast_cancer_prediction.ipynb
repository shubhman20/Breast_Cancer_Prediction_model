```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#load data
url = '/data.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
#return sum of null values
df.isna().sum()
```

```python
#this function will drop the column having null values
df = df.dropna(axis=1)
```

```python
#describe the dataset
df.describe()
```

```python
#get the count of malignant and benign
df['diagnosis'].value_counts()
```

```python
#label encoding(converting malignant to 1 and benign to 0 in diagnosis column)
from sklearn.preprocessing import LabelEncoder
labelencoder_Y = LabelEncoder()
df.iloc[:,1] = labelencoder_Y.fit_transform(df.iloc[:,1].values)
```

```python
#finding correlation between attributes
df.iloc[:,1:32].corr()
```

```python
#visualising the data
plt.figure(figsize=(15,15))
sns.heatmap(df.iloc[:,1:32].corr(),annot=True,fmt = ".0%")
```

```python
#splitting the dataset into dependent and independent variables
x = df.iloc[:,2:31].values
y = df.iloc[:,1].values
```

```python
#splitting values for training and testing purpose
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#feature scaling(z-score normaliization)
from sklearn.preprocessing import StandardScaler
X_train = StandardScaler().fit_transform(X_train)
X_test = StandardScaler().fit_transform(X_test)
```

```python
#defining logistic regression model
def model(X_train,Y_train):
  #Logistic regression
  from sklearn.linear_model import LogisticRegression
  log = LogisticRegression(random_state=0)
  log.fit(X_train,Y_train)
  print("Logistic Regression model accuracy: ",log.score(X_train,Y_train))
  return log
```

```python
model = model(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
print(classification_report(Y_test,model.predict(X_test)))
print("Accuracy: ",accuracy_score(Y_test,model.predict(X_test)))
```
